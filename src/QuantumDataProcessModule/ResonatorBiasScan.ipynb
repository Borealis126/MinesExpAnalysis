```python
import qdpm
import numpy as np
import matplotlib.pyplot as plt
```

```python
exp = qdpm.Experiment(r'O:\68707\BFXLD2222\LabVIEWData\Cooldown20211015\Heterodyne\Resonator Bias Scan\2021_11_02_00_15_40_incomplete')
```

```python
fig, ax = plt.subplots(figsize=(12, 10))

img = ax.pcolormesh(exp.mean_phase()[0], cmap='coolwarm')
ax.set_yticks(np.arange(0, exp.scan_size[0], 100))
ax.set_yticklabels(np.round(exp.scan_value[0][0::100], decimals=6))
ax.set_xticks(np.arange(0, 37, 6))
ax.set_xticklabels(np.round(exp.scan_value[1][0:37:6]/1000, decimals=2))
ax.set_xlabel('Bias (V)', fontsize=14)
ax.set_ylabel('Frequency (GHz)', fontsize=14)
cb = fig.colorbar(img)

ax.set_title('S21 phase. Cavity freq. vs Bias scan. Heterodyne system.')

plt.show()
fig.savefig(exp.path.joinpath(f'S21_phase_raw.png'), bbox_inches='tight')
```

```python
bg_subtracted = exp.mean_phase()[0].T-exp.mean_phase()[0].T.mean(axis=0)
fig, ax = plt.subplots(figsize=(12, 10))

img = ax.pcolormesh(bg_subtracted.T, cmap='coolwarm')
ax.set_yticks(np.arange(0, exp.scan_size[0], 100))
ax.set_yticklabels(np.round(exp.scan_value[0][0::100], decimals=6))
ax.set_xticks(np.arange(0, 37, 6))
ax.set_xticklabels(np.round(exp.scan_value[1][0:37:6]/1000, decimals=2))
ax.set_xlabel('Bias (V)', fontsize=14)
ax.set_ylabel('Frequency (GHz)', fontsize=14)
cb = fig.colorbar(img)

ax.set_title('S21 phase. Cavity freq. vs Bias scan. Heterodyne system.')

plt.show()
fig.savefig(exp.path.joinpath(f'S21_phase_bg_removed.png'), bbox_inches='tight')
```
