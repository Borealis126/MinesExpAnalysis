```python
import qdpm
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
```

```python
exp = qdpm.Experiment(r'O:\68707\BF2222\LABVIEWdata\Cooldown20210517\Heterodyne\Power scan\R2\2021_05_26_16_06_47')
```

```python
norm = np.zeros(exp.mean_mag()[0].T.shape)
for i in range(exp.mean_mag()[0].T.shape[0]):
    norm[i, :] = exp.mean_mag()[0].T[i, :]-(exp.scan_value[1][i]+15)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
img = ax.pcolormesh(norm[:, :], cmap='seismic')
# img = ax.pcolormesh(exp.mean_mag()[0].T, cmap='jet')
ax.set_xticks(np.arange(0, exp.scan_size[0], 40))
ax.set_xticklabels(np.round(exp.scan_value[0][0::40], decimals=6))
ax.set_yticks(np.arange(0, exp.scan_size[1], 2))
ax.set_yticklabels(np.round(exp.scan_value[1][:][0::2], decimals=2))
ax.set_xlabel('Frequency (GHz)', fontsize=14)
ax.set_ylabel('Power (dBm)', fontsize=14)
ax.set_title('Power Scan S21', fontsize=14)
ax.axvline(124, linestyle='--', c='r')
ax.axvline(58, linestyle='--', c='r')
ax.annotate('5.10166 GHz', xy=(133, 30))
ax.annotate('5.10016 GHz', xy=(5, 30))
cb = fig.colorbar(img)
plt.show()
```
