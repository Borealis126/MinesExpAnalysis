```python
import qdpm
import numpy as np
import matplotlib.pyplot as plt
```

```python
relaxation_exp = qdpm.Experiment(r'O:\68707\BF2222\LABVIEWdata\Cooldown20210716_Nb_SWIPHT_MB_W8_180nm\Heterodyne\Relaxation\01 -- 00\2021_07_21_15_03_23')
```

```python
qubit_idx = 1
```

```python
tau = np.linspace(0, 100, relaxation_exp.scan_size[0])
```

```python
clf = qdpm.SingleQubitClassifier(relaxation_exp, qubit=qubit_idx)
prediction = clf.predict()
population = 1 - prediction.mean(axis=1)
```

```python
fig, ax = plt.subplots()
relaxation_fit = qdpm.T1Fit(tau, population, make_plots=True, ax=ax)
ax.set_xlim(tau[0], tau[-1])
ax.set_ylim(0, 1)
fig.savefig(relaxation_exp.path.joinpath(f'Relaxation fit Q{qubit_idx}.png'), bbox_inches='tight')
plt.show()
```
